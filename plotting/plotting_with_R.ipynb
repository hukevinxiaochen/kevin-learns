## Lattice

### Problem

One way to learn whether the final A-level score (the variable `score`) depends on the average GCSE score (`gcsescore`) is to ask a slightly different question: is the distribution of `gcsescore` different for different values of `score`?

```r
pl <- histogram(~gcsescore | factor(score), data = Chem97)
print(pl)
```

#### The histogram() call

`histogram(~gcsescore | factor(score), data = Chem97)`

The first unnamed argument is a "formula" object.

`~ gcsescore | factor(score)`

`factor(score)`, the part following the vertical bar symbol, is the _conditioning variable_, indicating that the resulting plot should contain **one panel for each of its unique values** (levels).

## ggplot2

```r
pg <- ggplot(Chem97, aes(gcsescore)) + geom_histogram(binwidth = 0.5) + facet_wrap(~score)
print(pg)
```
